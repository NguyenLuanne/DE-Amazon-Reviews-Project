# Amazon Reviews EDA

## Retrieve data

```python
import json
import os
```

```python
def read_jsonl(URL: str) -> list:
    
    # Raise error if cannot find file path
    if not os.path.exists(URL):
        raise SyntaxError('File path not found')
    
    data = []
    with open(URL, 'r') as file:
        for line in file:
            data.append(json.loads(line))
            
    return data
```

```python
def show_data(data) -> None:
    print("Data type:", type(data))
    print("\n")
    print("First 5 values: \n")
    for value in data[:5]:
        print(json.dumps(value, indent=2))
```

### Subscription boxes

```python
subscription_boxes_reviews_URL = '../data/review_categories/Subscription_Boxes.jsonl'
subscription_boxes_reviews_data = read_jsonl(subscription_boxes_reviews_URL)
show_data(subscription_boxes_reviews_data)
```

```python
subscription_boxes_meta_URL = '../data/meta_categories/meta_Subscription_Boxes.jsonl'
subscription_boxes_meta_data = read_jsonl(subscription_boxes_meta_URL)
show_data(subscription_boxes_meta_data)
```
