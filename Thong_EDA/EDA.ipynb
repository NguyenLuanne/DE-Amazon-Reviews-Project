```python
import numpy as np
import pandas as pd
```

```python
reviews_data = pd.read_parquet('../data/reviews_data.parquet')
reviews_data.head()
```

```python
meta_data = pd.read_parquet('../data/meta_data.parquet')
meta_data.head()
```

```python
reviews_data.info()
```

```python
meta_data.info()
```

```python
null_categories = meta_data[meta_data['main_category'].isna()]
null_categories
```

```python
# Check if gift cards is the main category
null_categories['categories'].apply(lambda x: True if 'Gift Cards' in x else False).value_counts()
```

```python
meta_data['main_category'] = meta_data['main_category'].fillna('Gift Cards')
meta_data.info()
```

```python
main_category_values = meta_data['main_category'].value_counts()
main_category_outliers = main_category_values[main_category_values < 10].index.tolist()
```

```python
meta_data[meta_data['main_category'].isin(main_category_outliers)]
```
