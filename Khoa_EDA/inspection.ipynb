```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize= (10, 6))
sns.countplot(x='rating', data=reviews_df)
```

```python
reviews_df[~reviews_df['text'].isna()]
```

```python
review_length = reviews_df[(~reviews_df['text'].isna())]['text'].apply(len)
review_length
```

```python
reviews_df['review_length'] = reviews_df[~reviews_df['text'].isna()]['text'].apply(len)
plt.figure(figsize= (10,6))
sns.histplot(reviews_df['review_length'], bins= 20)
```

```python
#Sử dụng np.linspace để tạo ra các khoảng giá trị
reviews_length_bins = np.linspace(min(reviews_df['review_length']), max(reviews_df['review_length']), 4)

#Tạo một list tên các mốc giá trị
group_names = ['Short', 'Medium', 'Long']

#Gán nhãn các giá trị của biến 'horsepower' theo các mốc giá trị
reviews_df['reviews_length_bins'] = pd.cut(reviews_df['review_length'], reviews_length_bins, labels = group_names, include_lowest= True)
```

```python
for value in reviews_length_bins:
    print(f"{value:.2f}")
```

```python
plt.figure(figsize= (10,6))
plt.title("Biểu đồ biểu thị độ dài của các câu review")
sns.histplot(reviews_df['reviews_length_bins'],)
```

```python
reviews_df[~reviews_df['verified_purchase']]
```

```python
plt.figure(figsize= (15,8))

plt.subplot(1, 2, 1)
plt.title("Biểu đồ biểu thị độ dài của các câu review 5s nhưng ko mua hàng")
sns.histplot(reviews_df[(~reviews_df['verified_purchase']) & (reviews_df['rating'] == 5.0)]['reviews_length_bins'],)

plt.subplot(1, 2, 2)
plt.title("Biểu đồ biểu thị độ dài của các câu review 5s và mua hàng")
sns.histplot(reviews_df[(reviews_df['verified_purchase']) & (reviews_df['rating'] == 5.0)]['reviews_length_bins'],)
```
